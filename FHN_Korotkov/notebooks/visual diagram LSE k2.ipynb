```julia
k2_start = 0.0
k2_end = 100.0
len = 2000
range_param = range(k2_start, k2_end, length = len)
```

```julia
pathtoload = "/home/sergey/MEGA/dynamical-systems/FHN_Korotkov/data/"
LSE_s = (load(pathtoload * "LSEs_k2_length_2000_for_direction_1.jld2"))["LSEs"];
u0s = (load(pathtoload * "u0s_k2_length_2000_for_direction_1.jld2"))["u0ss"];
```

```julia
CairoMakie.activate!()
markersize = 1.5;
lbsize = 30;
ticksize = 30;
fig = Figure(size = (1200, 350))
axis = Axis(fig[1,1],
xlabel = L"k_2",  ylabel = L"Λs",
xlabelsize = lbsize, ylabelsize = lbsize,
xticklabelsize = ticksize,yticklabelsize = ticksize, xticks = [0, 20, 40, 60, 80, 100],
 xgridvisible = false, ygridvisible = false)

start_ = 1;end_ = 2000

lines!(axis, range_param[start_:end_],  LSE_s[start_:end_, 1], color = :red)
lines!(axis, range_param[start_:end_], LSE_s[start_:end_, 2], color = :green)
lines!(axis, range_param[start_:end_], LSE_s[start_:end_, 3], color = :blue)

#ylims!(-0.05, 0.05)
display(fig)
```
