```julia
k1_start = 0.094
k1_end = 0.0
len = 1000
range_param = range(k1_start, k1_end, length = len)
```

```julia
pathtoload = "/home/sergey/MEGA/dynamical-systems/FHN_Korotkov/data/"
LSE_s = (load(pathtoload * "LSEs_k1_from_0_90_to_0.jld2"))["LSEs"];
u0s = (load(pathtoload * "u0s_k1_from_0_90_to_0.jld2"))["u0ss"];
```

```julia
CairoMakie.activate!()
markersize = 1.5;
lbsize = 30;
ticksize = 30;
fig = Figure(size = (1200, 350))
axis = Axis(fig[1,1],
xlabel = L"k_2",  ylabel = L"x_1",
xlabelsize = lbsize, ylabelsize = lbsize,
xticklabelsize = ticksize,yticklabelsize = ticksize, xgridvisible = false, ygridvisible = false)

start_ = 1;end_ = 1000

lines!(axis, range_param[start_:end_],  LSE_s[start_:end_, 1], color = :red)
lines!(axis, range_param[start_:end_], LSE_s[start_:end_, 2], color = :green)
lines!(axis, range_param[start_:end_], LSE_s[start_:end_, 3], color = :blue)

display(fig)
```
