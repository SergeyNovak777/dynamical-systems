```julia
k1_start = 0.090
k1_end = 0.0
len = 1000
rangek1 = range(k1_start, k1_end, length = len)
```

```julia
pathtoload = "/home/sergey/MEGA/dynamical-systems/FHN_Korotkov/data/"
namefileload = "bifurcation_diagram_k1_from_0_90_to_0.jld2"

pathtosave = "/home/sergey/MEGA/dynamical-systems/FHN_Korotkov/images/"
namefilesave = "bifurcation_diagram_k1_from_0_90_to_0.pdf"

output = (load(pathtoload * namefileload))["output"]
```

```julia
CairoMakie.activate!()
markersize = 1.5;
lbsize = 30;
ticksize = 30;
fig = Figure(size = (1200, 350))
axis = Axis(fig[1,1],
xlabel = L"k_1",  ylabel = L"x_1",
xlabelsize = lbsize, ylabelsize = lbsize,
xticklabelsize = ticksize,yticklabelsize = ticksize, xgridvisible = false, ygridvisible = false)

for (j, p) in enumerate(rangek1)
    scatter!(axis, fill(p, length(output[j])), output[j];
        color = ("green", 0.5), markersize = markersize)
end
ylims!(-3.0, 2.0)
display(fig)
```

```julia
save(pathtosave * namefilesave, fig)
```
