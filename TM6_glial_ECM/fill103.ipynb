```julia
function y(x, c)
    return x.^2 ./ (c .+ x)
end
```

```julia
xsrC1 = range(-0.9, 10.0, length = 10000);
xslC1 = range(-10.0, -1.1, length = 10000);

xsrC2 = range(-99.9, 200.0, length = 10000);
xslC2 = range(-200.0, -100.1, length = 10000);

xsrC3 = range(-999.9, 1500.0, length = 10000);
xslC3 = range(-1500.0, -1000.1, length = 10000);
```

```julia
lw = 1.5;
f = Figure(resolution = (400, 400))
ax = Axis(f[1, 1], xlabel = L"x", ylabel = L"y", xlabelsize = 35, ylabelsize = 35, xticksize = 20, yticksize = 20, title = L"Solution $y(x)$", titlesize = 30)

lines!(xsrC1, y(xsrC1, 1.0), linewidth = lw, color = :black, label = L"C = 1")
lines!(xslC1, y(xslC1, 1.0), linewidth = lw, color = :black)
axislegend(ax, position = :rb)

xlims!(-5, 5)
ylims!(-7, 7)
display(f)
```
