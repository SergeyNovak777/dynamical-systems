```julia
Λs = load("LSs 500 change γY.jld")["data"]
Λs_1000 = load("LSs 1000 change γY to 50.0.jld")["data"]
Λs_2000 = load("LSs 2000 change γY to 100.0.jld")["data"]

num_spikes = load("num_spieks 500 change γY.jld")["data"]
num_spikes_1000 = load("num_spieks 1000 change γY.jld")["data"]
num_spikes_2000 = load("num_spieks 2000 change γY to 100.0.jld")["data"]

name_control_param = L"γY"
```

```julia
lb_size = 60; tck_size = 40
widthline = 2.5
```

```julia
start_param = 0.0
end_param = 1.0
length_range = 501
param_range = range(start_param, end_param, length = length_range)

st, ed = 1, 501
f = Figure(resolution = (2000, 500))
axis1 = Axis(f[1, 1], xlabel = name_control_param, ylabel = L"\Lambda s",
            xlabelsize = lb_size, ylabelsize = lb_size,
            xticklabelsize = tck_size, yticklabelsize = tck_size,
            xgridvisible = false, ygridvisible = false)

lines!(axis1, param_range[st:ed], Λs[st:ed, 1], linewidth = widthline, color = :red)
lines!(axis1, param_range[st:ed], Λs[st:ed, 2], linewidth = widthline, color = :green)
lines!(axis1, param_range[st:ed], Λs[st:ed, 3], linewidth = widthline, color = :blue)
hlines!(axis1, 0.0, linestyle = :dash, color = :black, linewidth = 0.5)

display(f)
save("LSE_0_1.eps", f)
```
