```julia
using StaticArrays, DifferentialEquations, DynamicalSystems
using CairoMakie, GLMakie
```

```julia
time = 500.0; tt = 250.0; tstep = 0.001; times = [time, tt]
integ_set = (alg = Vern9(), adaptive = false, dt = tstep);
```

```julia
cd("/home/sergey/MEGA/dynamical-systems/rate_model/images")
```

```julia
p = rate_model_get_params()
p[3] = 0.252 # s0E index 3
# s0I index 6

firstname = "Timeseries_s0E=$(p[3])_" 
```

```julia
p[21] = 0.5; # Y
u0 = [0.17952875685139708, 0.46521843882668207, 0.0062606073935254575, 0.06499657210006439, 1.2117657773513101e-18]
ds = CoupledODEs(rate_model, u0, p, diffeq = integ_set)
tr, trange = trajectory(ds, 1000, t = tstep; Ttr = tt)


tstart, tend = 700000, length(tr)
tickssize = 40
labelsize = 60
lw = 2.0

CairoMakie.activate!()
f= Figure(size = (2000, 500))
axisrHz  = Axis(f[1, 1], xlabel = L"time", ylabel = L"r[Hz]", xlabelsize = labelsize, ylabelsize = labelsize, xticklabelsize = tickssize, yticklabelsize = tickssize, xgridvisible = false, ygridvisible = false)

lines!(axisrHz, trange[tstart:tend], tr[tstart:tend, 3], linewidth = lw, color = :red, label = L"r_E")
lines!(axisrHz, trange[tstart:tend], tr[tstart:tend, 4], linewidth = lw, color = :blue, label = L"r_I")
axislegend(axisrHz, labelsize = 60)
display(f)

lastname = "gammaY=$(p[21]).eps"
save(firstname*lastname, f)
```
