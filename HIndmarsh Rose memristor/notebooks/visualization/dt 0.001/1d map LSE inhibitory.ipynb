# chemical inhibitory change k1_me

```julia
# inhibitory change  k_1me
s = load("LSs 300 inhibitory change k1_me.jld")["data"];
k1me_start = 0.0
k1me_end = 1.0
k1me_length = 300
st, ed = 1, 300
k1me_range = range(k1me_start, k1me_end, length = k1me_length)
lb_size = 30; tck_size = 25;
f = Figure(resolution = (1200, 300))
axis = Axis(f[1, 1], xlabel = L"k_{1 memr}", ylabel = L"\Lambda s",
            xlabelsize = lb_size, ylabelsize = lb_size,
            xticklabelsize = tck_size, yticklabelsize = tck_size,
            xgridvisible = false, ygridvisible = false, title = L"Inhibitory", titlesize = lb_size)

lines!(axis, k1me_range[st:ed], s[st:ed, 7], linewidth = 2.5, color = :red)
lines!(axis, k1me_range[st:ed], s[st:ed, 6], linewidth = 2.5, color = :orange)
lines!(axis, k1me_range[st:ed], s[st:ed, 5], linewidth = 2.5, color = :green)
lines!(axis, k1me_range[st:ed], s[st:ed, 4], linewidth = 2.5, color = :blue)

vlines!(axis, k1me_range[34], color = :deeppink, linestyle = :dash, linewidth = 2.5)
vlines!(axis, k1me_range[61], color = :deeppink, linestyle = :dash, linewidth = 2.5)
vlines!(axis, k1me_range[100], color = :deeppink, linestyle = :dash, linewidth = 2.5)

hlines!(axis, 0.0, linestyle = :dash, color = :black, linewidth = 2.0)
f
```
