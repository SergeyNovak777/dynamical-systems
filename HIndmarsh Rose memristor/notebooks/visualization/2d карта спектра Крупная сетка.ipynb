# неправильная система

```julia
Λs_50 = load("LSs_50x50.jld")["data"];
k1_merange = range( 0.0,  1.0, length = 51 )
k2_merange = range( 0.0,  1.0, length = 51)

index = 1
st, ed = 1, 51
absmax = maximum((Λs_50[st:ed, st:ed, index]))
mn, mx =  -absmax, absmax

f = Figure()
ax1 = Axis(f[1, 1], xlabel = L"k1_{me}",ylabel = L"k2_{me}", xlabelsize = 50, ylabelsize = 50,
            xticklabelsize = 35, yticklabelsize = 35,
            xgridvisible  = false, ygridvisible = false)

hm = heatmap!(ax1, k1_merange[st:ed], k2_merange[st:ed], Λs_50[st:ed, st:ed, index], colormap = :seismic, colorrange = (mn, mx))
Colorbar(f[1, 2], hm)
f
```
