```julia
Λs = load("LSs 100.jld")["data"];


k2me_start = 0.0
k2me_end = 1.0
k2me_length = 101

k2me_range = range(k2me_start, k2me_end, length = k2me_length)

lb_size = 30; tck_size = 25;

f = Figure(resolution = (1200, 300))
axis = Axis(f[1, 1], xlabel = L"k_{2 memr}", ylabel = L"\Lambda s",
            xlabelsize = lb_size, ylabelsize = lb_size,
            xticklabelsize = tck_size, yticklabelsize = tck_size,
            xgridvisible = false, ygridvisible = false)

lines!(axis, k2me_range, Λs[:, 1], linewidth = 2.5, color = :red)
lines!(axis, k2me_range, Λs[:, 2], linewidth = 2.5, color = :orange)
lines!(axis, k2me_range, Λs[:, 3], linewidth = 2.5, color = :green)

f
```
