```julia
# excitatory change  k_1me
Λs = load("LSs 300 excitatory change k1_me to 1.jld")["data"];
u0s = load("u0s 300 excitatory change k1_me to 1.jld")["data"];

k1me_start = 0.0
k1me_end = 1.0
k1me_length = 300
k1me_range = range(k1me_start, k1me_end, length = k1me_length)

st, ed = 1, 300
lb_size = 40; tck_size = 35;
f = Figure(resolution = (1200, 400))
axis = Axis(f[1, 1], xlabel = L"k_{1}", ylabel = L"\Lambda s",
            xlabelsize = lb_size, ylabelsize = lb_size,
            xticklabelsize = tck_size, yticklabelsize = tck_size,
            xgridvisible = false, ygridvisible = false) #, title = L"Chaos", titlesize = lb_size)

lines!(axis, k1me_range[st:ed], Λs[st:ed, 7], linewidth = 2.5, color = :red)
lines!(axis, k1me_range[st:ed], Λs[st:ed, 6], linewidth = 2.5, color = :orange)
lines!(axis, k1me_range[st:ed], Λs[st:ed, 5], linewidth = 2.5, color = :green)
lines!(axis, k1me_range[st:ed], Λs[st:ed, 4], linewidth = 2.5, color = :blue)

vlines!(axis, k1me_range[15], color = :deeppink, linewidth = 2.5)
vlines!(axis, k1me_range[150], color = :deeppink, linewidth = 2.5)
vlines!(axis, k1me_range[77], color = :deeppink, linewidth = 2.5)

hlines!(axis, 0.0, linestyle = :dash, color = :black, linewidth = 2.0)
f
```
