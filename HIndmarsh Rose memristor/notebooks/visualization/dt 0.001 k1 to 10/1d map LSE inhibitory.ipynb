```julia
# inhibitory change  k_1me
s = load("LSs 300 inhibitory change k1_me to 1.jld")["data"];
Es = load("EEs 300 inhibitory change k1_me to 1.jld")["data"];
u0s = load("u0s 300 inhibitory change k1_me to 1.jld")["data"];

k1me_start = 0.0
k1me_end = 1.0
k1me_length = 299
st, ed = 1, 299
k1me_range = range(k1me_start, k1me_end, length = k1me_length)
lb_size = 30; tck_size = 25;
f = Figure(resolution = (1200, 400))
axis1 = Axis(f[1, 1], xlabel = L"k_{1}", ylabel = L"\Lambda s",
            xlabelsize = lb_size, ylabelsize = lb_size,
            xticklabelsize = tck_size, yticklabelsize = tck_size,
            xgridvisible = false, ygridvisible = false) #, title = L"Inhibitory", titlesize = lb_size)

#axis2 = Axis(f[2, 1], xlabel = L"k_{1 memr}", ylabel = L"\Lambda s",
            #xlabelsize = lb_size, ylabelsize = lb_size,
            #xticklabelsize = tck_size, yticklabelsize = tck_size,
            #xgridvisible = false, ygridvisible = false, title = L"Inhibitory", titlesize = lb_size)

lines!(axis1, k1me_range[st:ed], s[st:ed, 7], linewidth = 2.5, color = :red)
lines!(axis1, k1me_range[st:ed], s[st:ed, 6], linewidth = 2.5, color = :orange)
lines!(axis1, k1me_range[st:ed], s[st:ed, 5], linewidth = 2.5, color = :green)
lines!(axis1, k1me_range[st:ed], s[st:ed, 4], linewidth = 2.5, color = :blue)
hlines!(axis1, 0.0, linestyle = :dash, color = :black, linewidth = 0.5)

lines!(axis1, k1me_range[st:ed], s[st:ed, 4], linewidth = 2.5, color = :blue)

vlines!(axis1, k1me_range[30], color = :deeppink, linestyle = :dash, linewidth = 2.5)
vlines!(axis1, k1me_range[62], color = :deeppink, linestyle = :dash, linewidth = 2.5)
vlines!(axis1, k1me_range[100], color = :deeppink, linestyle = :dash, linewidth = 2.5)


#lines!(axis2, k1me_range[st:ed], Es[st:ed], linewidth = 2.5, color = :deeppink)

f
```
