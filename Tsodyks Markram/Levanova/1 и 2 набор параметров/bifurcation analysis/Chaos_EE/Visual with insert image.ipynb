```julia
path = "C:\\Users\\Alex\\Desktop\\repo\\2612\\branch\\dynamical-systems\\Tsodyks Markram\\Levanova\\bifurcation analysis\\Chaos_EE\\Total\\"
path1 = "C:\\Users\\Alex\\Desktop\\repo\\2612\\branch\\dynamical-systems\\Tsodyks Markram\\Levanova\\bifurcation analysis\\Chaos_EE\\left space\\"

path = path * "E_many_plots.jld"
path1 = path1 * "E_left_space_5000_dots.jld";
```

```julia
I0range_total = range( -1.5094, -1.5115, length = 5000 )
I0range_insert = range( -1.51075, -1.51090, length = 5000 );
```

```julia
E_total = load(path)["data"]
E_left =  load(path1)["data"];
```

```julia
f = Figure(resolution = (1200, 700))
axis = Axis(f[1,1]; xlabel = L"I_0", ylabel = L"E", xlabelsize = 35, ylabelsize = 35,
    xticklabelsize = 30, yticklabelsize = 30,
    xgridvisible = false, ygridvisible = false)

ax2 = Axis(f, bbox = BBox(140, 450, 200, 350), xticklabelsize = 13,
    yticklabelsize = 13,
    xgridvisible = false, ygridvisible = false)

for (j, p) in enumerate(I0range_total)
    scatter!(axis, fill(p, length(E_total[j])), E_total[j]; color = ("blue", 0.5), markersize = 2.5)
end
for (j, p) in enumerate(I0range_insert)
    scatter!(ax2, fill(p, length(E_left[j])), E_left[j]; color = ("blue", 0.5), markersize = 2.5)
end
f
```
