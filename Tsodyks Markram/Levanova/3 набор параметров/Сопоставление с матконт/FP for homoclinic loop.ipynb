```julia
using StaticArrays, DifferentialEquations, DynamicalSystems, LinearAlgebra
E, x, y  = 0..40, 0..1,  0..1
box = E × x × y
using CairoMakie, GLMakie;
```

```julia
const τ = 0.013;  const τD = 0.0799364822793578;  const τy = 3.3;  const J = 3.07;  const β = 0.300
const xthr = 0.75; const ythr = 0.4
const α = 1.58; const ΔU0 = 0.305; 
```

```julia
time = -100.0; tt = 0; tstep = 0.001
integ_set = (alg = Feagin12(), adaptive = false, dt = tstep)
```

```julia
U0 = 0.265;
I0 = -1.7062964892514
p =  [α, τ, τD, τy, J, xthr, ythr, U0, ΔU0, β, I0]
u0 = [8.344960582744857, 0.7384956468337779, 0.43830335836164663]
```

```julia
prob = ODEProblem(TM, u0, (0.0, 100.0), p)
solforward = solve(prob, alg = Feagin12(), adaptive = false, dt = 0.001)
```

```julia
prob = ODEProblem(TM, u0, (0.0, -100.0), p)
solback = solve(prob, alg = Feagin12(), adaptive = false, dt = 0.001)
```

```julia
tsback, tfback = 1, 6500
tsfor, tffor = 1, 100000

f = Figure(resolution = (700, 700))
ax = LScene(f[1, 1], show_axis = true)

scale!(ax.scene, 50, 50, 1)

lines!(solback[3, tsback:tfback], solback[2, tsback:tfback], solback[1, tsback:tfback], linewidth = 1.0, color = :black)
lines!(solforward[3, tsfor:tffor], solforward[2, tsfor:tffor], solforward[1, tsfor:tffor], linewidth = 1.0, color = :deeppink)

display(GLMakie.Screen(), f)
```
