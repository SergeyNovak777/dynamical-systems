```julia
Λs= load("LSs_80x80_down_up.jld")["data"];
```

```julia
I0range = range( -0.5, -2.0, length = 80 )
U0range = range( 0.22,  0.55, length = 80 )

index = 1
absmax = maximum(abs.(Λs[:, :, index]))

mn, mx =  -absmax, absmax

f = Figure()
ax1 = Axis(f[1, 1], xlabel = L"I0",ylabel = L"U0", xlabelsize = 50, ylabelsize = 50,
            xticklabelsize = 35, yticklabelsize = 35,
            xgridvisible  = false, ygridvisible = false)

hm = heatmap!(ax1, I0range, U0range, Λs[:, :, index], colormap = :seismic, colorrange = (mn, mx))
Colorbar(f[1, 2], hm)
f
```
