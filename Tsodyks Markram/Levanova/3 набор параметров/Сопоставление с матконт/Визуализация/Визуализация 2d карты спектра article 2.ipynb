```julia
Λs= load("LSs_300_300_article2.jld")["data"];
u0s = load("u0s_300_300_article2.jld")["data"];
```

```julia
I0range = range( -1.5, -2.0,  length = 300 )
U0range = range( 0.4,  0.26, length = 300 )

index = 1
I0st, I0ed = 1, 300
U0st, U0ed = 1, 300
absmax = maximum(abs.(Λs[I0st:I0ed, U0st:U0ed, index]))

mn, mx =  -absmax, absmax
CairoMakie.activate!()
f = Figure()

ax1 = Axis(f[1, 1], xlabel = L"I0",ylabel = L"u_0", xlabelsize = 50, ylabelsize = 50,
            xticklabelsize = 35, yticklabelsize = 35,
            xgridvisible  = false, ygridvisible = false)

hm = heatmap!(ax1, I0range[I0st:I0ed], U0range[U0st:U0ed], Λs[I0st:I0ed, U0st:U0ed, index], colormap = :seismic,
                 colorrange = (mn, mx))
Colorbar(f[1, 2], hm)
display(f)
```
