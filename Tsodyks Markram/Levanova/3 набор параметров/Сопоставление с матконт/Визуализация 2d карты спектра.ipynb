```julia
using JLD, CairoMakie
```

```julia
Λs_90 = load("LSs_90_90_time_400.jld")["data"];
Λs_100 = load("LSs_100_100.jld")["data"];
```

```julia
I0range = range( -0.8, -2.8, length = 90 )
U0range = range( 0.22,  0.6, length = 90 )

index = 1
absmax = maximum((Λs_90[:, :, index]))
mn, mx =  -absmax, absmax

f = Figure()
ax1 = Axis(f[1, 1], xlabel = L"I0",ylabel = L"U0", xlabelsize = 50, ylabelsize = 50,
            xticklabelsize = 35, yticklabelsize = 35,
            xgridvisible  = false, ygridvisible = false)

hm = heatmap!(ax1, I0range, U0range, Λs_90[:, :, index], colormap = :seismic, colorrange = (mn, mx))
Colorbar(f[1, 2], hm)
f
```
