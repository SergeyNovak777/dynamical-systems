```julia
using Pkg
Pkg.activate("D:\\Science\\Code\\Julia\\envr")
```

```julia
using JLD, CairoMakie
```

```julia
I0range = range( -1.40, -1.709, length = 100 )
U0range = range( 0.3,  0.47, length = 100)
```

```julia
Λs = load("LS_100x100.jld")["data"];
```

```julia
absmax = maximum(abs.(Λs[:, :, 1]))
mn, mx =  -absmax, absmax
```

```julia
f = Figure()
ax1 = Axis(f[1, 1], xlabel = L"I0",ylabel = L"U0", xlabelsize = 50, ylabelsize = 50,
            xticklabelsize = 35, yticklabelsize = 35,
            xgridvisible  = false, ygridvisible = false)

hm = heatmap!(ax1, I0range, U0range, Λs[:, :, 1], colormap = :seismic, colorrange = (mn, mx))
Colorbar(f[1, 2], hm)
f
```
