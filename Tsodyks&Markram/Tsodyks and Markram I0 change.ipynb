```julia
using DynamicalSystems
using DifferentialEquations
using StaticArrays


using CairoMakie
```

```julia
U(y) = U0 + ΔU0 / ( 1 + exp(-50*(y-ythr)) )
σ(x) = 1 / ( 1 + exp(-20*(x-xthr)) )
xthr = .7; ythr = .6
```

```julia
function model(u, p, t)
    
    
    E, x, y = u
    α, τ, τD, τy, J, U0, ΔU0, β, I0 = p
    
    
    du1 = (-E + α * log( 1 + exp( (J * U(y) * x * E + I0  ) / (α) ) ) ) / τ
    du2 = (1-x)/τD - U(y)*x*E
    du3 = (-y)/τy+β*σ(x)
    
    return SVector(du1, du2, du3)
end
```

```julia
τ = 0.013; τD = 0.15; τy = 1.8;
α = 1.15
J = 3.07
β = 0.4475
U0 = .3; ΔU0 = .305
I0 = -0.011

p = SA[α, τ, τD, τy, J, U0, ΔU0, β, I0]
tspan = (0.0, 2000.0)
```

```julia
minE = 0; maxE = 17;

ts, tf = 500000, 520000
```

## α = 1.15; I0 = -0.011

```julia
u0 = SA[1.11,   0.988967 ,  0.4861];

α = 1.15
I0 = -0.011
p = SA[α, τ, τD, τy, J, U0, ΔU0, β, I0]

prob = ODEProblem(model, u0, tspan, p)
sol = solve(prob, RK4(),  adaptive = false, dt = 0.001);
```

```julia
f = Figure(resolution = (1200, 270))

ax1 = Axis(f[1, 1], xlabel = "t", ylabel = "E, X, Y", xlabelsize = 21, ylabelsize = 21,xticklabelsize = 15, yticklabelsize = 15)


lines!(ax1, sol[1,ts:tf], color = "blue", label = "E")
lines!(ax1, sol[2,ts:tf], color = "orange", label = "X")
lines!(ax1, sol[3,ts:tf], color = "green", label = "Y")

axislegend(ax1, labelsize=15)
f
```

```julia
function labeldetect(cord)
    
    if cord == 1
        label = "E"
    elseif cord == 2
        label = "X"
    elseif cord == 3
        label = "Y"
    end
    return label
end
```
