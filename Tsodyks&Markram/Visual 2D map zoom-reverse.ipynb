```julia
function parabola(x)
    a = 0.13625841750841752
    b = 0.189709595959596
    c = 0.22
    return  a * (x.^2) + b * x .+ c 
end
```

```julia
forward = readdlm("2d_map_parabola_zoom.txt");
backward = readdlm("2d_map_parabola_zoom_reverse.txt"); println(size(backward)[1])
total = vcat(forward, backward);
```
